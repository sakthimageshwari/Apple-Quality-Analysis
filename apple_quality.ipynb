# Splitting dataset for Training and Testing

```python
from sklearn.model_selection import train_test_split 
```

```python
x = ma.iloc[:,ma.columns!='Quality'] #data
y = ma.iloc[:,ma.columns=='Quality'] #outcome or Label
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2)
```

```python
x
```

```python
y
```

# Importing Algorithm

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
model=RandomForestClassifier()
```

```python
model.fit(xtrain,ytrain.values.ravel())
```

```python
predict_output = model.predict(xtest) #to test the alg
print(predict_output)
```

# Accuracy Prediction

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(predict_output,ytest)
print('The accuracy score for RF:',acc)
```
